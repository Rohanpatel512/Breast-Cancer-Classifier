## Import Libraries

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
```

## Data Preprocessing

```python
# Load in the dataset
cancer_dataset = pd.read_csv('breast-cancer.csv')

# Show the first five rows
cancer_dataset.head()
```

```python
# Display information for each of the features
cancer_dataset.describe()
```

```python
# Check for missing values in each column
cancer_dataset.isnull().sum()
```

```python
# Encode the target feature (diagnosis) (0 = Benign, 1 = Malignant)
cancer_dataset['diagnosis'] = cancer_dataset['diagnosis'].map(lambda x: 0 if x == 'B' else 1)
cancer_dataset.head()
```

```python
# Remove 'id' column from the dataset
cancer_dataset.drop('id', inplace=True, axis=1)
# Get the correlation between target feature and each of the descriptive features
cancer_dataset.corr(method='pearson', numeric_only=True)['diagnosis']
```

```python
# Visualize the number of malignant and benign cases (check for imbalance)
malignant_cases = len(cancer_dataset[cancer_dataset['diagnosis'] == 1])
benign_cases = len(cancer_dataset[cancer_dataset['diagnosis'] == 0])

cases = [malignant_cases, benign_cases]
label = ["Malignant", "Benign"]
plt.pie(cases,  autopct='%1.1f%%', labels=label)
plt.legend()
plt.show()
```

```python
# Visualize the relationship between radius_mean and diagnosis
figure = plt.figure(figsize=(13, 10))
plt.subplot(2, 2, 1)
sns.barplot(x=cancer_dataset['diagnosis'], y=cancer_dataset['radius_mean'], errorbar=None)
plt.title("Diagnosis vs Radius Mean")
plt.xlabel("Diagnosis")
plt.ylabel("Radius Mean")

plt.subplot(2, 2, 2)
sns.barplot(x=cancer_dataset['diagnosis'], y=cancer_dataset['perimeter_mean'])
plt.title("Diagnosis vs Perimeter Mean")
plt.xlabel("Diagnosis")
plt.ylabel("Perimeter Mean")

plt.subplot(2, 2, 3)
sns.barplot(x=cancer_dataset['diagnosis'], y=cancer_dataset['concave points_mean'])
plt.title("Diagnosis vs Concave points mean")
plt.xlabel("Diagnosis")
plt.ylabel("Concave Point Mean")

plt.subplot(2, 2, 4)
sns.barplot(x=cancer_dataset['diagnosis'], y=cancer_dataset['concave points_worst'])
plt.title("Diagnosis vs Concave Points worst")
plt.xlabel("Diagnosis")
plt.ylabel("Concave Point Worst")
```

```python
correlations = cancer_dataset.corr().abs()
plt.figure(figsize=(20, 20))
mask = np.triu(np.ones_like(correlations, dtype=bool))
sns.heatmap(correlations, mask=mask, annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

```python
# Remove features causing multicollinearity 
features = ['area_worst', 'perimeter_worst', 'radius_worst', 'area_mean', 'perimeter_mean']
cancer_dataset.drop(features, inplace=True, axis=1)
```

```python
columns = cancer_dataset.columns 
x = cancer_dataset[columns[1:]]
y = cancer_dataset['diagnosis']

# Split the dataset into 60% training and 40% testing 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)

# Standardize the descriptive features 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

## Model Training

```python
model = SVC()
model.fit(x_train_scaled, y_train)
```
